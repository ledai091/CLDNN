```python
import cv2

# Path to the video file
video_path = '1.mp4'

# Open the video
cap = cv2.VideoCapture(video_path)

# Check if video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Frame rate of the video
frame_rate = cap.get(cv2.CAP_PROP_FPS)

# Interval of frames to capture
interval = int(frame_rate)  # Capturing an image every second

count = 0
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()

    # If read was successful
    if ret:
        # Save frame every 'interval' frames
        if frame_count % interval == 0:
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            cv2.imwrite(f'img/frame_{count}.jpg', gray)
            count += 1
        frame_count += 1
    else:
        break

# Release the video capture object
cap.release()
cv2.destroyAllWindows()
```

```python
def crop_center(img):
    height, width = img.shape[0], img.shape[1]
    mid_y = height//2
    mid_x = width//2
    img_resize = img[(mid_x-(mid_x//2)):(mid_x+(mid_x//2)), (mid_y-(mid_y//2)):(mid_y+(mid_y//2))]
    return img_resize

im = crop_center(image)
plt.imshow(im)
```

```python
import csv
def dict_to_csv(data, filename):
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Key', 'Value'])
        for key, value in data.items():
            writer.writerow([key, value])
            
dict_to_csv(dict, 'mapping.csv')
```
